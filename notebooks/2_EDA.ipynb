```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load  dataset 
cleaned_data = pd.read_csv("../Data/cleaned_ecommerce_data.csv")
```

```python
# Price distribution
plt.figure(figsize=(8, 6))
sns.histplot(cleaned_data['price'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Set the order for the review scores
review_order = [5.0, 4.0, 3.0, 2.0, 1.0]

# Review Score Distribution with custom order
plt.figure(figsize=(8, 6))
sns.countplot(x='review_score', data=cleaned_data, order=review_order)
plt.title('Review Score Distribution')
plt.xlabel('Review Score')
plt.ylabel('Count')
plt.show()
```

```python
# Top 10 product categories by number of orders
top_categories = cleaned_data['product_category_name_english'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title('Top 10 Product Categories by Number of Orders')
plt.xlabel('Product Category')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# Price vs Freight Value
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=cleaned_data,
    x='price', 
    y='freight_value',
    alpha=0.6
)
plt.title('Price vs. Shipping Cost (BRL)', fontsize=14)
plt.xlabel('Product Price (R$)', fontsize=12)
plt.ylabel('Freight Value (R$)', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Top 10 products by number of orders
top_products = cleaned_data['product_id'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top 10 Products by Number of Orders')
plt.xlabel('Product ID')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 1: Calculate Delivery Time
cleaned_data['order_purchase_timestamp'] = pd.to_datetime(cleaned_data['order_purchase_timestamp'])
cleaned_data['order_delivered_customer_date'] = pd.to_datetime(cleaned_data['order_delivered_customer_date'])
cleaned_data['delivery_time'] = (cleaned_data['order_delivered_customer_date'] - cleaned_data['order_purchase_timestamp']).dt.days


plt.figure(figsize=(10,6))
sns.histplot(cleaned_data['delivery_time'], kde=True, bins=50)
plt.title('Distribution of Delivery Time')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Frequency')
plt.show()
```
